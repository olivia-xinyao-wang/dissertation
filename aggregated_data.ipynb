```python
#Housing 
def housing_data(file_path,year):
    df = pd.read_csv(file_path)
    df[['county', 'state']] = df['Geographic Area Name'].str.split(', ', expand=True)
    
    df = df[['county', 'state','Estimate!!Median monthly housing costs']]
    df= df.rename(columns={'Estimate!!Median monthly housing costs':'Median_monthly_housing_costs'})
    df['year']=str(year)
    return df
```

```python
housing_usa_2020 = housing_data('E://dissertation//Housing_cost_us//2020Data.csv',2020)
housing_usa_2019 = housing_data('E://dissertation//Housing_cost_us//2019Data.csv',2019)
housing_usa_2018 = housing_data('E://dissertation//Housing_cost_us//2018Data.csv',2018)
housing_usa_2017 = housing_data('E://dissertation//Housing_cost_us//2017Data.csv',2017)
housing_usa_2016 = housing_data('E://dissertation//Housing_cost_us//2016Data.csv',2016)
housing_usa_2015 = housing_data('E://dissertation//Housing_cost_us//2015Data.csv',2015)
housing_usa_2014 = housing_data('E://dissertation//Housing_cost_us//2014Data.csv',2014)
housing_usa_2013 = housing_data('E://dissertation//Housing_cost_us//2013Data.csv',2013)
housing_usa_2012 = housing_data('E://dissertation//Housing_cost_us//2012Data.csv',2012)
housing_usa_2011 = housing_data('E://dissertation//Housing_cost_us//2011Data.csv',2011)
```

```python
import pandas as pd
from linearmodels.panel import PanelOLS
import statsmodels.api as sm

#model  = PanelOLS.from_formula('status_score ~ Household_Income + giniindex + Household_Size + Marital_Status + Age_And_Presence_Of_Children + Male_Head_Employment+Median_monthly_housing_costs+ Population + EntityEffects', data = filtered_df )


# 定义自变量（包括常数项）
X = filtered_df[['Household_Income', 'giniindex', 'Household_Size', 'Marital_Status', 'Age_And_Presence_Of_Children','Male_Head_Employment',	
                 'Female_Head_Employment', 
                 'Median_monthly_housing_costs', 'Population']]
X = sm.add_constant(X)  # 添加常数项

# 定义因变量
y = filtered_df['status_score']

# 拟合固定效应模型（within 模型）
model = PanelOLS(y, X, entity_effects=True)
results = model.fit(cov_type='clustered', cluster_entity=True)

# 输出模型摘要
print(results.summary)
```
